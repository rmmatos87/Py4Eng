# Básico

## Criando tela em branco
```python
fig, ax = plt.subplots()
```
* `fig` é uma instância de figura (”quadro em branco”)
* `ax` é um frame no qual você pode plotar algo

````python
fig, ax = plt.subplots()
plt.show()
````

## Plotando uma curva
$$
\{x \in \mathbb{R}\ |\ 0 \le x \le 10\} \\
y = \sin{x}
$$
`len(x)= 200`

````python
x = np.linspace(0, 10, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(
    x, y,
    'b-', # b: linha azul;
          # -: estilo da linha;
    linewidth=5 # espessura da linha
)
plt.show()
````

## Plotando duas curvas
$$
\{x \in \mathbb{R}\ |\ 0 \le x \le 10\} \\
y = \sin{x}
z = \cos(x)
$$
`len(x)= 200`

````python
x = np.linspace(0, 10, 200)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots()
ax.plot(x, y, 'b-', linewidth=5) # plotando o seno
ax.plot(x, z, 'r--', linewidth=3) # plotando o cosseno
plt.show()
````

As representações utilizando o matplotlib são bem customizáveis considerando-se os métodos disponíveis na biblioteca.

## Rótulos e Legendas
* `plt.subplots`
    * `figsize` - `tuple` - ajusta o tamanho da figura (largura, altura)
* `ax.plot`
    * `label` - `str` - nomeia a curva
* `plt.xlabel` - `str` - nomeia o eixo X
* `plt.ylabel` - `str` - nomeia o eixo Y
* [`ax.legend`](https://matplotlib.org/stable/api/legend_api.html?highlight=legend#module-matplotlib.legend) - insere a legenda


````python
x = np.linspace(0, 10, 200)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'b-', linewidth=5, label='y = sin(x)', alpha=.6) # plotando o seno
ax.plot(x, z, 'r--', linewidth=3, label='y = cos(x)') # plotando o cosseno

plt.xlabel('X')
plt.ylabel('Y')

ax.legend(loc='upper center') # mostra a legenda com a localização definida

plt.show()
````

## Utilizando [LaTeX](https://www.math.ubc.ca/~pwalls/math-python/jupyter/latex/)

````python
x = np.linspace(0, 10, 200)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'b-', linewidth=5, label='$y = \sin{(x)}$', alpha=.6) # Utilizando LaTeX
ax.plot(x, z, 'r--', linewidth=3, label='y = cos(x)') # plotando o cosseno

plt.xlabel('X')
plt.ylabel('Y')

ax.legend(loc='upper center') # mostra a legenda com a localização definida

plt.show()
````

* `ax.set_title` ou `plt.title` - str - Insere um título no gráfico
* `ax.set_yticks` ou `plt.yticks` - list - Escolhe quais as marcações no eixo Y

````python
x = np.linspace(0, 10, 200)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'b-', linewidth=5, label='$y = \sin{(x)}$', alpha=.6) # Utilizando LaTeX
ax.plot(x, z, 'r--', linewidth=3, label='y = cos(x)') # plotando o cosseno

plt.xlabel('X')
plt.ylabel('Y')

ax.legend(loc='upper center') # mostra a legenda com a localização definida

ax.set_title('Teste')
ax.set_yticks(range(-1, 2))

plt.show()
````

## Gráficos descontínuos
Para funções que apresentem descontinuidades podemos representar as descontinuidades com diferentes cores:
$$y = \bigg\{ \begin{matrix} -x^2 & x \le 0 \\ x^2 + 1 & x > 0 \end{matrix}$$

````python
x = np.linspace(-2, 2, 200)
y = np.zeros_like(x)
y[x <= 0] = -x[x <= 0]**2
y[x > 0] = x[x <= 0]**2 + 1

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'x-')

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
````

Para eliminar a conexão entre os pontos de descontinuidades inserimos um $NaN$ (*Not a Number*). Sabemos que a derivada da função aumenta quando ocorre a discontinuidade, portanto vamos observar a derivada discreta da função.

* `np.diff` - `np.ndarray` - faz a diferenças entre os elementos subsequentes do vetor

````python
x = np.linspace(-2, 2, 200)
y = np.zeros_like(x)
y[x <= 0] = -x[x <= 0]**2
y[x > 0] = x[x <= 0]**2 + 1

dydx = np.diff(y)/np.diff(x) # retorna uma dimensão a menos que o vetor original

fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(x[1:], dydx, 'r3-', label='$\Delta{y}/\Delta{x}$')

ax.legend()

plt.show()
````

Um modo simples de encontrar uma descontinuidade pode ser encontrar a maior dervada. Neste ponto será adicionado o `np.nan` para mostrar a descontinuidade.
```python
np.insert(
    arr,      # array onde o valor será inserido
    obj,      # será inserido antes das posições definidas por este iterável ou inteiro
    values,   # valor a ser inserido
    axis=None # eixo para a inserção
)
```

````python
x = np.linspace(-2, 2, 200)
y = np.zeros_like(x)
y[x <= 0] = -x[x <= 0]**2
y[x > 0] = x[x <= 0]**2 + 1

dydx = np.diff(y)/np.diff(x) # retorna uma dimensão a menos que o vetor original

descontinuidade = np.where(dydx == dydx.max())[0]

x = np.insert(x, descontinuidade + 1, np.nan)
y = np.insert(y, descontinuidade + 1, np.nan)

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(x, y, 'x-')

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
````

# Histogramas
---

## Histograma de uma distrinuição normal

````python
n_pontos = 1000
y = np.random.randn(n_pontos)

fig, ax = plt.subplots()

n_bins = 30 # número de colunas

ax.hist(y, bins=n_bins) # gerando o histograma

plt.show()
````

## Subplots
Subplots são matrizes de plots em suas linhas e colunas.
```python
plt.subplot(
    nrows=1,          # número de gráficos na horizontal
    ncols=1,          # número de gráficos na vertical
    sharex=False,     # utilizar o mesmo eixo - X
    sharey=False,     # utilizar o mesmo eixo - Y
    squeeze=True,     # modifica a saída de eixos para um vetor achatado (1-D)
    subplot_kw=None,  # argumentos de matplotlib.pyplot.subplot
    gridspec_kw=None, # argumentos de matplotlib.pyplot.grid
    **fig_kw          # argumentos de matplotlib.pyplot.figure
)
```

````python
n_pontos = 1000
y = np.random.randn(n_pontos)

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(20,5))

n_bins = 30 # número de colunas

ax[0].hist(y, bins=n_bins)
ax[0].set_title('Histograma com %d colunas'%n_bins)
ax[1].hist(y, bins=20, facecolor='r')
ax[1].set_title('Histograma com 20 colunas')

plt.show()
````

Três medidores de nível (`LT-01`, `LT-02`, `LT-03`) são instalados em um tanque. Cada um deles tem um princípio de funcionamento diferente. Também são diferentes os erros de medição.
* Crie uma função que plote as 3 medições de nível no intervalo de tempo $t_0 \rightarrow t_1$ - `plotar_niveis(t_0, t_1)`
* Disponibilize uma entrada para função que controle se o resultado irá ser plotado em um só gráfico ou em três subplots em linha - `plotar_niveis(t_0, t_1, subplots=False)`
* Gere `boxplots` para os dados em uma mesma escala de "y" compare o desempenho dos pedidores em termo de precisão e exatidão. Observe que o período mais transitório pode poluir o boxplot, tente removê-lo!

````python
t = np.linspace(0, 100, 1000) # tempo de 0 a 100
nivel_inicial = 50 # nível em 50%
instante = 9

nivel_real = np.ones_like(t)*nivel_inicial 
nivel_real[t > instante] = 50 + 5*np.exp(-(t[t > instante] - instante)/5)*np.cos(np.pi*(t[t > instante])/12)

def medidor(valor, erro=0, bias=0):
    return valor + erro*np.random.randn(len(valor)) + bias*valor

LT01 = medidor(nivel_real, .05, .01)
LT02 = medidor(nivel_real, .10, -.01)
LT03 = medidor(nivel_real, .15)

plt.subplots(figsize = (15, 5))
plt.plot(t, LT01, '3',
         t, LT02, '3',
         t, LT03, '3',
         t, nivel_real, '--',
         alpha=.6)
plt.ylabel('Nível [%]')
plt.xlabel('t [min]')
plt.legend(['LT-01', 'LT-02', 'LT-03', 'Valor real'])
plt.show()
````
